### 전국대학학과정보 데이터 분석

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\hhj53\Desktop\전국대학별학과정보표준데이터.csv", encoding='cp949')

df.head()
```

```python
for name in df["학과명"].unique().tolist():
    print(name)
```

```python
from collections import Counter
import re

# 모든 학과명을 띄어쓰기 기준으로 분해한 단어 리스트 생성
all_words = []
for name in df["학과명"].dropna():
    words = re.findall(r'[가-힣A-Za-z0-9]+', name)
    all_words.extend(words)

# 가장 많이 등장한 단어 100개 보기
common_words = Counter(all_words).most_common(100)

# 출력
for word, count in common_words:
    print(f"{word}: {count}")
```

```python
# 키워드 기반 관련 학과 필터링
keywords = [
    "컴퓨터", "소프트웨어", "정보보호", "사이버", "보안", "AI", "인공지능", 
    "데이터", "ICT", "IT", "정보통신", "디지털", "네트워크", 
    "빅데이터", "사물인터넷", "클라우드", "스마트", "블록체인", "융합"
]
related = df[df["학과명"].str.contains('|'.join(keywords), case=False, na=False)]
```

```python
related
```

```python
# 지역별 학과 수 집계
region_counts = related.groupby("시도명")["학교명"].nunique().reset_index()
region_counts.columns = ["시도", "관련학과대학수"]
print(region_counts)
```

```python
top5 = region_counts.sort_values("관련학과대학수", ascending=False).head(5)
print(top5)
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# Windows용 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False  
```

```python
plt.figure(figsize=(12, 6))
plt.bar(region_counts["시도"], region_counts["관련학과대학수"])
plt.xticks(rotation=45)
plt.title("시도별 사이버보안/IT 관련 학과 보유 대학 수")
plt.ylabel("대학 수")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### 부산 사이버보안 컨퍼런스이므로 영남권 지역 재분석

```python
# 영남권 시도 설정
yeongnam = ["부산광역시", "대구광역시", "울산광역시", "경상남도", "경상북도"]

# 관련 학과 + 영남권 필터링
related_yeongnam = related[related["시도명"].isin(yeongnam)]

# 시군구명 기준 집계
by_cityname = related_yeongnam.groupby(["시도명", "시군구명"])["학교명"].nunique().reset_index()
by_cityname.columns = ["시도", "시군구", "관련학과대학수"]
by_cityname = by_cityname.sort_values("관련학과대학수", ascending=False)
print(by_cityname.head(10))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 7))
sns.barplot(data=by_cityname.head(10), x="관련학과대학수", y="시군구", hue="시도", dodge=False)
plt.title("영남권 시군구별 사이버보안/IT 관련 학과 대학 수 (Top 10)")
plt.xlabel("관련학과 대학 수")
plt.ylabel("시군구")
plt.tight_layout()
plt.show()
```

```python
# 관련 학과 수 기준 대학별 집계
target_univs = related_yeongnam.groupby(["시도명", "시군구명", "학교명"])["학과명"].nunique().reset_index()
target_univs.columns = ["시도", "시군구", "학교명", "관련학과수"]

# 관련학과수 기준 상위 대학 정렬
top_targets = target_univs.sort_values("관련학과수", ascending=False)

# 상위 10개 대학
print(top_targets.head(10))
```

```python
# 1. 쉼표 제거 + 숫자로 변환
df_경산["2022년_계_연령구간인구수"] = df_경산["2022년_계_연령구간인구수"].replace(",", "", regex=True).astype(float)
df_경산["2023년_계_연령구간인구수"] = df_경산["2023년_계_연령구간인구수"].replace(",", "", regex=True).astype(float)

# 2. 청년인구 합산
df_경산["청년인구"] = df_경산["2022년_계_연령구간인구수"] + df_경산["2023년_계_연령구간인구수"]

# 3. 경산시만 필터링
df_경산_filtered = df_경산[df_경산["행정구역"].str.contains("경상북도 경산시")]

# 4. 결과 출력
df_경산_result = df_경산_filtered[["행정구역", "청년인구"]].reset_index(drop=True)
display(df_경산_result)
```

```python
# 1. 경산 결과에서 0번 행만 추출
df_경산_요약 = df_경산_result.iloc[[0]].copy()

# 2. 부산 5구 + 경산 합치기
df_통합 = pd.concat([df_청년5구_filtered, df_경산_요약], ignore_index=True)

# 3. 결과 출력
display(df_통합)
```

#### 점수화(관련대학학과수 + 지역청년수)

```python
# 1. 관련학과수와 행정구역을 키로 병합 준비
top_targets["행정구역"] = top_targets["시도"] + " " + top_targets["시군구"]

# 2. 대학별 청년인구 결합
df_scored = pd.merge(top_targets, df_통합, on="행정구역", how="left")

# 3. 결측치 제거
df_scored = df_scored.dropna(subset=["청년인구"])

# 4. 정규화
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# 괄호 제거하여 병합 키 통일
df_통합["행정구역"] = df_통합["행정구역"].str.extract(r"([^\(]+)").iloc[:, 0].str.strip()
top_targets["행정구역"] = top_targets["시도"].str.strip() + " " + top_targets["시군구"].str.strip()

df_scored = pd.merge(top_targets, df_통합, on="행정구역", how="left")
df_scored = df_scored.dropna(subset=["청년인구"])


df_scored["관련학과수_정규화"] = scaler.fit_transform(df_scored[["관련학과수"]])
df_scored["청년인구_정규화"] = scaler.fit_transform(df_scored[["청년인구"]])

# 5. 점수 계산
alpha, beta = 0.5, 0.5
df_scored["홍보우선순위점수"] = df_scored["관련학과수_정규화"] * alpha + df_scored["청년인구_정규화"] * beta

# 6. 점수 기준 정렬
df_ranked = df_scored.sort_values("홍보우선순위점수", ascending=False).reset_index(drop=True)

# 7. 최종 출력
df_ranked_final = df_ranked[["학교명", "시도", "시군구", "관련학과수", "청년인구", "홍보우선순위점수"]]
display(df_ranked_final)
```

#### 분석 해석
청년인구와 관련학과 수를 균등 반영한 점수 기준으로

부산진구 동의대학교는 청년도 많고 관련학과도 많아 최우선 홍보 대상

경산 소재 대학들(대구대·경일대 등)도 높은 점수를 보임 → 영남권 외곽 집중 가능성

점수가 낮은 학교는 청년인구 또는 학과수가 적음 → 우선순위 후순위 가능
